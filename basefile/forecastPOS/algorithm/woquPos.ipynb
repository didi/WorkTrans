```python
import time,datetime

#21/3/2019 05:00:04

def getT(time_original):
    time_format = datetime.datetime.strptime(time_original, '%d/%m/%Y %H:%M:%S')
    time_format = time_format.strftime('%Y-%m-%d %H:%M:%S')
    return  time_format[0:10]
getT('21/3/2019 05:00:04')  
```

```python
data['t']=data.bill_year.map(str)+'-'+data.bill_month.map(str)+'-'+data.bill_day.map(str)
```

```python
dataOne=data[data.did==6]
dataOne['tb']=dataOne.gmt_bill.map(getT)
dataOne=dataOne[['tb','bill_hour','bill_minute','data_value','peoples']]
dataOne.shape
```
